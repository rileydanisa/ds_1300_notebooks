```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
facts_largest = facts.sort_values(by='Pop_2020',ascending=False)
facts_largest.head(5)
```

```python
#more_water = facts[facts['Area_water']>facts['Area_land']]
#more_water.sort_values(by=['Area_water'], ascending=False)
facts[facts['Area_water']>facts['Area_land']].sort_values(by=['Area_water'], ascending=False)
```

```python
facts['Pop_growth'] = facts.apply(lambda x: x['Pop_2020']-x['Pop_2010'],axis=1)
facts[['State', 'Pop_growth']].sort_values(by='Pop_growth',ascending=False).head(1)
```

```python
large_bird=facts[['Pop_2020','State_bird']].groupby(by='State_bird').sum()
large_bird.sort_values(by=['Pop_2020'], ascending=False).head(5)
```

```python
area_bird=facts[['Area_land','State_bird']].groupby(by='State_bird').sum()
area_bird.sort_values(by=['Area_land'], ascending=False).head(5)
```

```python
facts['State'][facts['Capital_pop']== facts['city_1_pop']].count()
```

```python
facts['P_drop']=facts.apply(lambda x: 100 * (x['city_1_pop']-x['city_5_pop'])/x['city_1_pop'], axis=1)
facts['City_1'][facts['P_drop'] == facts['P_drop'].max()]
```
